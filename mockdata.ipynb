```python
import pandas as pd
from pathlib import Path
from datetime import datetime
```

```python
transaction_df = pd.read_csv(str(Path.cwd())+'/data/Daily Household Transactions.csv')
transaction_df.head()
```

```python
# assume currency exchange rate 1 INR to 0.0092 pounds
transaction_df['Amount(Pounds)'] = transaction_df['Amount'] * 0.0092
transaction_df['Datetime'] = pd.to_datetime(transaction_df['Date'], format='%d/%m/%Y %H:%M:%S', errors='coerce')
transaction_df['Datetime'].fillna(pd.to_datetime(transaction_df['Date'], format='%d/%m/%Y', errors='coerce'), inplace=True)
transaction_df['date'] = transaction_df['Datetime'].apply(lambda x: x.date())
transaction_df.head()
transaction_df.drop(columns=['Currency', 'Amount','Date', 'Datetime'], inplace=True)
```

```python
transaction_df['date'] = transaction_df['date'].apply(lambda x: x.replace(month=8))
```

```python
transaction_df
```

```python
transaction_df['date'].dropna(inplace=True)

import pandas as pd

# Original date as a string
latest_date = '20/8/2018'

# Convert the string to a datetime object
latest_date = pd.to_datetime(latest_date, dayfirst=True)

# Subtract 3 months using pd.DateOffset
three_months_ago = latest_date - pd.DateOffset(months=3)


# Filter the DataFrame for transactions from the last 3 months
latest_month_data = transaction_df[transaction_df['date'] >= three_months_ago.date()]

# Show the first few rows of the filtered data
latest_month_data
```
